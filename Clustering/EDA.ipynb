```python
import os
from dotenv import load_dotenv
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from skimpy import skim
from IPython.display import display
pd.set_option('display.max_columns', None)

load_dotenv()
data_dir = os.getenv('DATA_DIR', 'data')
warnings.filterwarnings('ignore')
```

```python
dataset=os.path.join(data_dir, 'clusteringmidterm.csv')
assert os.path.exists(dataset), f"Dataset not found at {dataset}"
```

```python
df=pd.read_csv(dataset)
display(skim(df))
```

## 🔍 Missing Values Analysis

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Column': missing_values.index,
    'Missing_Count': missing_values.values,
    'Missing_Percentage': missing_percent.values
})
missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)

if len(missing_df) > 0:
    print("⚠️  MISSING VALUES DETECTED:")
    print("="*80)
    display(missing_df)
    
    # Visualize missing values
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.barh(missing_df['Column'], missing_df['Missing_Percentage'], color='coral')
    ax.set_xlabel('Missing Percentage (%)', fontweight='bold')
    ax.set_title('Missing Values by Column', fontsize=14, fontweight='bold')
    ax.grid(alpha=0.3, axis='x')
    plt.tight_layout()
    plt.show()
else:
    print("✅ No missing values detected in the dataset!")
```

```python
#handle missing values if any
df=df.dropna()
print("\nAfter handling missing values:")
print(f"Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
```

```python
# Statistical summary of numerical columns
print("="*80)
print("STATISTICAL SUMMARY - NUMERICAL FEATURES")
print("="*80)
display(df.describe().T.style.background_gradient(cmap='YlOrRd', subset=['mean', 'std']))
```

## 🔢 Duplicate Records Check

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
duplicate_percent = (duplicates / len(df)) * 100

print("="*80)
print("DUPLICATE RECORDS")
print("="*80)
print(f"Total duplicate rows: {duplicates:,} ({duplicate_percent:.2f}%)")

if duplicates > 0:
    print(f"\n⚠️  Action: Removing {duplicates:,} duplicate rows...")
    df_original_shape = df.shape
    df = df.drop_duplicates()
    print(f"✅ Dataset shape: {df_original_shape} → {df.shape}")
else:
    print("✅ No duplicate rows found!")
```

## 📊 Feature Distributions

```python
# Select only numerical columns for distribution analysis
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

print(f"Numerical features: {len(numerical_cols)}")
print(f"Features: {numerical_cols}\n")

# Plot distributions for all numerical features
n_cols = 4
n_rows = (len(numerical_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5 * n_rows))
axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes

for idx, col in enumerate(numerical_cols):
    ax = axes[idx]
    
    # Histogram with KDE
    ax.hist(df[col].dropna(), bins=50, alpha=0.7, color='skyblue', edgecolor='black')
    ax.set_title(f'{col}\n(Mean: {df[col].mean():.2f}, Std: {df[col].std():.2f})', 
                 fontweight='bold', fontsize=10)
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')
    ax.grid(alpha=0.3)

# Hide unused subplots
for idx in range(len(numerical_cols), len(axes)):
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
```

## 📦 Outlier Detection (Box Plots)

```python
# Box plots to detect outliers
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5 * n_rows))
axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes

for idx, col in enumerate(numerical_cols):
    ax = axes[idx]
    
    # Box plot
    bp = ax.boxplot(df[col].dropna(), vert=True, patch_artist=True)
    bp['boxes'][0].set_facecolor('lightblue')
    bp['boxes'][0].set_alpha(0.7)
    
    ax.set_title(f'{col}', fontweight='bold', fontsize=10)
    ax.set_ylabel('Value')
    ax.grid(alpha=0.3, axis='y')

# Hide unused subplots
for idx in range(len(numerical_cols), len(axes)):
    axes[idx].axis('off')

plt.tight_layout()
plt.show()

# Calculate outlier statistics using IQR method
print("\n" + "="*80)
print("OUTLIER STATISTICS (IQR Method)")
print("="*80)

outlier_summary = []
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_count = len(outliers)
    outlier_percent = (outlier_count / len(df)) * 100
    
    outlier_summary.append({
        'Feature': col,
        'Outlier_Count': outlier_count,
        'Outlier_Percentage': f"{outlier_percent:.2f}%",
        'Lower_Bound': f"{lower_bound:.2f}",
        'Upper_Bound': f"{upper_bound:.2f}"
    })

outlier_df = pd.DataFrame(outlier_summary)
outlier_df = outlier_df[outlier_df['Outlier_Count'] > 0].sort_values('Outlier_Count', ascending=False)

if len(outlier_df) > 0:
    display(outlier_df)
else:
    print("✅ No outliers detected in any feature!")
```

## 🔗 Correlation Analysis

```python
# Correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Plot correlation heatmap
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Feature Correlation Matrix', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# Find highly correlated features
print("\n" + "="*80)
print("HIGHLY CORRELATED FEATURES (|correlation| > 0.7)")
print("="*80)

high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            high_corr_pairs.append({
                'Feature_1': correlation_matrix.columns[i],
                'Feature_2': correlation_matrix.columns[j],
                'Correlation': correlation_matrix.iloc[i, j]
            })

if high_corr_pairs:
    high_corr_df = pd.DataFrame(high_corr_pairs).sort_values('Correlation', key=abs, ascending=False)
    display(high_corr_df)
    print(f"\n⚠️  {len(high_corr_pairs)} highly correlated feature pairs found!")
    print("Consider removing redundant features to reduce multicollinearity.")
else:
    print("✅ No highly correlated features found (|correlation| > 0.7)")
```

## 🎯 Feature Scaling & Normalization

**Note:** We're scaling the FULL dataset (keeping outliers) as recommended for clustering.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Check if scaling is needed (different scales)
print("="*80)
print("FEATURE SCALES - BEFORE SCALING")
print("="*80)

scale_info = pd.DataFrame({
    'Feature': numerical_cols,
    'Min': [df[col].min() for col in numerical_cols],
    'Max': [df[col].max() for col in numerical_cols],
    'Range': [df[col].max() - df[col].min() for col in numerical_cols],
    'Mean': [df[col].mean() for col in numerical_cols],
    'Std': [df[col].std() for col in numerical_cols]
})
display(scale_info)

# Apply StandardScaler (recommended for clustering)
print("\n" + "="*80)
print("APPLYING STANDARD SCALER (Z-score normalization)")
print("="*80)
print("Formula: z = (x - mean) / std")
print("Result: Mean = 0, Std = 1 for all features")

scaler = StandardScaler()
df_scaled_standard = pd.DataFrame(
    scaler.fit_transform(df[numerical_cols]),
    columns=numerical_cols,
    index=df.index
)

print(f"\n✅ StandardScaler applied to {len(numerical_cols)} features")

# Also create MinMaxScaler version (alternative)
print("\n" + "="*80)
print("APPLYING MIN-MAX SCALER (0-1 normalization)")
print("="*80)
print("Formula: x_scaled = (x - min) / (max - min)")
print("Result: All values between 0 and 1")

minmax_scaler = MinMaxScaler()
df_scaled_minmax = pd.DataFrame(
    minmax_scaler.fit_transform(df[numerical_cols]),
    columns=numerical_cols,
    index=df.index
)

print(f"✅ MinMaxScaler applied to {len(numerical_cols)} features")

# Compare before and after
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Original data
axes[0].boxplot([df[col] for col in numerical_cols[:5]], labels=numerical_cols[:5])
axes[0].set_title('Original Data (First 5 Features)', fontweight='bold')
axes[0].set_ylabel('Value')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(alpha=0.3)

# StandardScaler
axes[1].boxplot([df_scaled_standard[col] for col in numerical_cols[:5]], labels=numerical_cols[:5])
axes[1].set_title('StandardScaler (Mean=0, Std=1)', fontweight='bold')
axes[1].set_ylabel('Scaled Value')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(alpha=0.3)

# MinMaxScaler
axes[2].boxplot([df_scaled_minmax[col] for col in numerical_cols[:5]], labels=numerical_cols[:5])
axes[2].set_title('MinMaxScaler (Range 0-1)', fontweight='bold')
axes[2].set_ylabel('Scaled Value')
axes[2].tick_params(axis='x', rotation=45)
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("SCALING SUMMARY")
print("="*80)
print(f"StandardScaler - Mean: {df_scaled_standard.mean().mean():.4f}, Std: {df_scaled_standard.std().mean():.4f}")
print(f"MinMaxScaler - Min: {df_scaled_minmax.min().min():.4f}, Max: {df_scaled_minmax.max().max():.4f}")
```
