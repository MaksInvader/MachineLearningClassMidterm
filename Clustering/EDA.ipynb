## 📦 Outlier Detection (Box Plots)

```python
# Box plots to detect outliers
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5 * n_rows))
axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes

for idx, col in enumerate(numerical_cols):
    ax = axes[idx]
    
    # Box plot
    bp = ax.boxplot(df[col].dropna(), vert=True, patch_artist=True)
    bp['boxes'][0].set_facecolor('lightblue')
    bp['boxes'][0].set_alpha(0.7)
    
    ax.set_title(f'{col}', fontweight='bold', fontsize=10)
    ax.set_ylabel('Value')
    ax.grid(alpha=0.3, axis='y')

# Hide unused subplots
for idx in range(len(numerical_cols), len(axes)):
    axes[idx].axis('off')

plt.tight_layout()
plt.show()

# Calculate outlier statistics using IQR method
print("\n" + "="*80)
print("OUTLIER STATISTICS (IQR Method)")
print("="*80)

outlier_summary = []
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_count = len(outliers)
    outlier_percent = (outlier_count / len(df)) * 100
    
    outlier_summary.append({
        'Feature': col,
        'Outlier_Count': outlier_count,
        'Outlier_Percentage': f"{outlier_percent:.2f}%",
        'Lower_Bound': f"{lower_bound:.2f}",
        'Upper_Bound': f"{upper_bound:.2f}"
    })

outlier_df = pd.DataFrame(outlier_summary)
outlier_df = outlier_df[outlier_df['Outlier_Count'] > 0].sort_values('Outlier_Count', ascending=False)

if len(outlier_df) > 0:
    display(outlier_df)
else:
    print("✅ No outliers detected in any feature!")
```

## 🛡️ Outlier & Anomaly Handling Strategy

**Important Decision for Clustering:**

For **customer segmentation clustering**, outliers can represent:
- ✅ **Valuable rare segments** (e.g., VIP customers, high spenders, unique behavior patterns)
- ❌ **Data errors or noise** (e.g., measurement errors, data entry mistakes)

**Our Approach:**
1. **Detect outliers** using multiple methods (IQR, Z-score, Isolation Forest)
2. **Analyze outlier characteristics** to understand if they're valuable or noise
3. **Create multiple datasets** with different outlier handling strategies
4. **Let clustering algorithms decide** (e.g., DBSCAN naturally handles outliers as noise)

**Recommendation:** Keep outliers for clustering unless they're clearly data errors!

```python
# Method 1: Z-Score Based Outlier Detection
from scipy import stats

print("="*80)
print("METHOD 1: Z-SCORE OUTLIER DETECTION")
print("="*80)
print("Rule: |Z-score| > 3 indicates an outlier (99.7% confidence)")
print()

z_scores = np.abs(stats.zscore(df[numerical_cols]))
z_outliers = (z_scores > 3).any(axis=1)
z_outlier_count = z_outliers.sum()

print(f"Total rows with Z-score outliers: {z_outlier_count:,} ({z_outlier_count/len(df)*100:.2f}%)")

# Count outliers per feature
z_outliers_per_feature = (np.abs(stats.zscore(df[numerical_cols])) > 3).sum()
z_outliers_df = pd.DataFrame({
    'Feature': numerical_cols,
    'Z_Score_Outliers': z_outliers_per_feature.values
}).sort_values('Z_Score_Outliers', ascending=False)

print("\nOutliers per feature (Z-score > 3):")
display(z_outliers_df[z_outliers_df['Z_Score_Outliers'] > 0])

# Visualize Z-score distribution
fig, ax = plt.subplots(figsize=(12, 6))
if z_outlier_count > 0:
    z_outliers_df_filtered = z_outliers_df[z_outliers_df['Z_Score_Outliers'] > 0]
    ax.barh(z_outliers_df_filtered['Feature'], z_outliers_df_filtered['Z_Score_Outliers'], color='coral', alpha=0.7)
    ax.set_xlabel('Number of Z-Score Outliers (|Z| > 3)', fontweight='bold')
    ax.set_title('Z-Score Outliers by Feature', fontsize=14, fontweight='bold')
    ax.grid(alpha=0.3, axis='x')
    plt.tight_layout()
    plt.show()
else:
    print("✅ No Z-score outliers detected!")
```

```python
# Method 2: Isolation Forest (ML-based anomaly detection)
from sklearn.ensemble import IsolationForest

print("\n" + "="*80)
print("METHOD 2: ISOLATION FOREST ANOMALY DETECTION")
print("="*80)
print("Advanced ML algorithm for detecting anomalies in high-dimensional data")
print("Contamination = 0.05 (assumes 5% of data are anomalies)")
print()

# Fit Isolation Forest
iso_forest = IsolationForest(
    contamination=0.05,  # Expected proportion of outliers
    random_state=42,
    n_jobs=-1
)

# Predict outliers (-1 = outlier, 1 = inlier)
iso_predictions = iso_forest.fit_predict(df[numerical_cols])
iso_outliers = iso_predictions == -1
iso_outlier_count = iso_outliers.sum()

print(f"Total anomalies detected: {iso_outlier_count:,} ({iso_outlier_count/len(df)*100:.2f}%)")

# Get anomaly scores
iso_scores = iso_forest.score_samples(df[numerical_cols])
df_temp = df.copy()
df_temp['anomaly_score'] = iso_scores
df_temp['is_anomaly'] = iso_outliers

print(f"\nAномaly score range: [{iso_scores.min():.4f}, {iso_scores.max():.4f}]")
print(f"Lower scores = more anomalous")

# Visualize anomaly scores
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Score distribution
axes[0].hist(iso_scores, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
axes[0].axvline(x=iso_scores[iso_outliers].max(), color='red', linestyle='--', 
                label=f'Anomaly Threshold (score < {iso_scores[iso_outliers].max():.3f})')
axes[0].set_xlabel('Anomaly Score', fontweight='bold')
axes[0].set_ylabel('Frequency', fontweight='bold')
axes[0].set_title('Isolation Forest Anomaly Score Distribution', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Anomaly vs Normal
labels = ['Normal', 'Anomaly']
counts = [(~iso_outliers).sum(), iso_outliers.sum()]
colors = ['#90EE90', '#FF6B6B']
axes[1].bar(labels, counts, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
axes[1].set_ylabel('Count', fontweight='bold')
axes[1].set_title('Normal vs Anomaly Classification', fontsize=14, fontweight='bold')
axes[1].grid(alpha=0.3, axis='y')

for i, (label, count) in enumerate(zip(labels, counts)):
    axes[1].text(i, count + len(df)*0.01, f'{count:,}\n({count/len(df)*100:.1f}%)', 
                ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\n✅ Isolation Forest model trained")
```

```python
# Method 3: Compare all outlier detection methods
print("\n" + "="*80)
print("OUTLIER DETECTION METHOD COMPARISON")
print("="*80)

# Create comparison DataFrame
outlier_comparison = pd.DataFrame({
    'Method': ['IQR (1.5×IQR)', 'Z-Score (|Z| > 3)', 'Isolation Forest (5%)'],
    'Outliers_Detected': [
        outlier_df['Outlier_Count'].astype(int).sum() if len(outlier_df) > 0 else 0,
        z_outlier_count,
        iso_outlier_count
    ],
    'Percentage': [
        f"{(outlier_df['Outlier_Count'].astype(int).sum() / len(df) * 100):.2f}%" if len(outlier_df) > 0 else "0.00%",
        f"{z_outlier_count/len(df)*100:.2f}%",
        f"{iso_outlier_count/len(df)*100:.2f}%"
    ]
})

display(outlier_comparison)

# Visualize comparison
fig, ax = plt.subplots(figsize=(10, 6))
methods = outlier_comparison['Method']
counts = outlier_comparison['Outliers_Detected']
colors = ['#FF6B6B', '#FFA07A', '#FFD700']

bars = ax.bar(methods, counts, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
ax.set_ylabel('Number of Outliers Detected', fontweight='bold')
ax.set_title('Outlier Detection Method Comparison', fontsize=14, fontweight='bold')
ax.grid(alpha=0.3, axis='y')

# Add value labels
for bar, count in zip(bars, counts):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(count):,}\n({count/len(df)*100:.1f}%)',
            ha='center', va='bottom', fontweight='bold')

plt.xticks(rotation=15, ha='right')
plt.tight_layout()
plt.show()

print("\n📊 Agreement between methods:")
print(f"   • All methods agree: {(z_outliers & iso_outliers).sum():,} samples")
print(f"   • Z-score only: {(z_outliers & ~iso_outliers).sum():,} samples")
print(f"   • Isolation Forest only: {(~z_outliers & iso_outliers).sum():,} samples")
```

```python
# Analyze outlier characteristics
print("\n" + "="*80)
print("OUTLIER CHARACTERISTICS ANALYSIS")
print("="*80)

if iso_outlier_count > 0:
    print("\n🔍 Comparing Normal vs Anomaly samples:\n")
    
    # Statistical comparison
    comparison_stats = pd.DataFrame()
    
    for col in numerical_cols[:5]:  # Show first 5 features for brevity
        comparison_stats[col] = [
            f"{df[~iso_outliers][col].mean():.2f}",
            f"{df[iso_outliers][col].mean():.2f}",
            f"{df[~iso_outliers][col].std():.2f}",
            f"{df[iso_outliers][col].std():.2f}"
        ]
    
    comparison_stats.index = ['Normal_Mean', 'Anomaly_Mean', 'Normal_Std', 'Anomaly_Std']
    
    display(comparison_stats.T)
    
    # Visualize feature distributions for normal vs anomalies
    n_features_to_plot = min(4, len(numerical_cols))
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    axes = axes.flatten()
    
    for idx, col in enumerate(numerical_cols[:n_features_to_plot]):
        ax = axes[idx]
        
        # Normal samples
        ax.hist(df[~iso_outliers][col], bins=30, alpha=0.6, label='Normal', color='skyblue', edgecolor='black')
        # Anomaly samples
        ax.hist(df[iso_outliers][col], bins=30, alpha=0.6, label='Anomaly', color='coral', edgecolor='black')
        
        ax.set_xlabel(col, fontweight='bold')
        ax.set_ylabel('Frequency', fontweight='bold')
        ax.set_title(f'{col} Distribution: Normal vs Anomaly', fontweight='bold')
        ax.legend()
        ax.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n💡 Insights:")
    print("   • Anomalies may represent extreme values or unusual patterns")
    print("   • For customer segmentation: anomalies could be VIP customers or rare behaviors")
    print("   • Decision: Keep them for clustering to discover these rare segments!")
else:
    print("✅ No significant anomalies detected")
```

## 🎯 Outlier Handling Decision & Strategy

**For Customer Segmentation Clustering:**

### ✅ RECOMMENDED: Keep Outliers
**Why?**
1. Outliers may represent valuable customer segments (VIP, high-value, unique behaviors)
2. DBSCAN naturally handles outliers by classifying them as "noise"
3. Other algorithms (K-Means, GMM) will create small clusters for outlier groups
4. You can always filter them out AFTER analyzing cluster results

### 📊 Alternative Datasets Created:
We'll create 3 versions of the dataset:
1. **Full dataset** (with outliers) - RECOMMENDED
2. **Z-score filtered** (without extreme Z-score outliers)
3. **Isolation Forest filtered** (without ML-detected anomalies)

This gives you flexibility to experiment with different approaches!

```python
# Create multiple datasets with different outlier handling strategies
print("="*80)
print("CREATING DATASETS WITH DIFFERENT OUTLIER STRATEGIES")
print("="*80)

# Strategy 1: Keep all data (RECOMMENDED for clustering)
df_full = df.copy()
print(f"\n1️⃣ Full Dataset (All Data - RECOMMENDED)")
print(f"   Shape: {df_full.shape}")
print(f"   Samples: {len(df_full):,}")
print(f"   ✅ Best for discovering all customer segments including rare ones")

# Strategy 2: Remove Z-score outliers (|Z| > 3)
df_no_zscore = df[~z_outliers].copy()
print(f"\n2️⃣ Z-Score Filtered Dataset (|Z| ≤ 3)")
print(f"   Shape: {df_no_zscore.shape}")
print(f"   Samples: {len(df_no_zscore):,}")
print(f"   Removed: {len(df) - len(df_no_zscore):,} samples ({(len(df) - len(df_no_zscore))/len(df)*100:.2f}%)")
print(f"   ⚠️  May lose valuable extreme customer profiles")

# Strategy 3: Remove Isolation Forest anomalies
df_no_iso = df[~iso_outliers].copy()
print(f"\n3️⃣ Isolation Forest Filtered Dataset")
print(f"   Shape: {df_no_iso.shape}")
print(f"   Samples: {len(df_no_iso):,}")
print(f"   Removed: {len(df) - len(df_no_iso):,} samples ({(len(df) - len(df_no_iso))/len(df)*100:.2f}%)")
print(f"   ⚠️  Removes ML-detected anomalies")

# Visualize dataset sizes
fig, ax = plt.subplots(figsize=(10, 6))

datasets = ['Full Dataset\n(RECOMMENDED)', 'Z-Score Filtered\n(|Z| ≤ 3)', 'Isolation Forest\nFiltered']
sizes = [len(df_full), len(df_no_zscore), len(df_no_iso)]
colors = ['#90EE90', '#FFD700', '#FFA07A']

bars = ax.bar(datasets, sizes, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
ax.set_ylabel('Number of Samples', fontweight='bold')
ax.set_title('Dataset Sizes with Different Outlier Strategies', fontsize=14, fontweight='bold')
ax.grid(alpha=0.3, axis='y')

# Add value labels
for bar, size in zip(bars, sizes):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{size:,}\n({size/len(df)*100:.1f}%)',
            ha='center', va='bottom', fontweight='bold')

# Add horizontal line for original size
ax.axhline(y=len(df), color='red', linestyle='--', alpha=0.5, label=f'Original: {len(df):,}')
ax.legend()

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("📊 RECOMMENDATION FOR CLUSTERING")
print("="*80)
print("✅ Use the FULL DATASET (Strategy 1)")
print()
print("Reasons:")
print("  • Outliers may represent valuable customer segments")
print("  • DBSCAN naturally handles outliers as 'noise' clusters")
print("  • K-Means will group similar outliers into small clusters")
print("  • You can analyze and decide AFTER seeing cluster results")
print("  • Preserves all information for business insights")
print()
print("⚠️  Alternative strategies available if needed for specific use cases")
```

## 🎯 Feature Scaling & Normalization

**Note:** We're scaling the FULL dataset (keeping outliers) as recommended for clustering.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Check if scaling is needed (different scales)
print("="*80)
print("FEATURE SCALES - BEFORE SCALING")
print("="*80)

scale_info = pd.DataFrame({
    'Feature': numerical_cols,
    'Min': [df[col].min() for col in numerical_cols],
    'Max': [df[col].max() for col in numerical_cols],
    'Range': [df[col].max() - df[col].min() for col in numerical_cols],
    'Mean': [df[col].mean() for col in numerical_cols],
    'Std': [df[col].std() for col in numerical_cols]
})
display(scale_info)

# Apply StandardScaler (recommended for clustering)
print("\n" + "="*80)
print("APPLYING STANDARD SCALER (Z-score normalization)")
print("="*80)
print("Formula: z = (x - mean) / std")
print("Result: Mean = 0, Std = 1 for all features")

scaler = StandardScaler()
df_scaled_standard = pd.DataFrame(
    scaler.fit_transform(df[numerical_cols]),
    columns=numerical_cols,
    index=df.index
)

print(f"\n✅ StandardScaler applied to {len(numerical_cols)} features")

# Also create MinMaxScaler version (alternative)
print("\n" + "="*80)
print("APPLYING MIN-MAX SCALER (0-1 normalization)")
print("="*80)
print("Formula: x_scaled = (x - min) / (max - min)")
print("Result: All values between 0 and 1")

minmax_scaler = MinMaxScaler()
df_scaled_minmax = pd.DataFrame(
    minmax_scaler.fit_transform(df[numerical_cols]),
    columns=numerical_cols,
    index=df.index
)

print(f"✅ MinMaxScaler applied to {len(numerical_cols)} features")

# Compare before and after
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Original data
axes[0].boxplot([df[col] for col in numerical_cols[:5]], labels=numerical_cols[:5])
axes[0].set_title('Original Data (First 5 Features)', fontweight='bold')
axes[0].set_ylabel('Value')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(alpha=0.3)

# StandardScaler
axes[1].boxplot([df_scaled_standard[col] for col in numerical_cols[:5]], labels=numerical_cols[:5])
axes[1].set_title('StandardScaler (Mean=0, Std=1)', fontweight='bold')
axes[1].set_ylabel('Scaled Value')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(alpha=0.3)

# MinMaxScaler
axes[2].boxplot([df_scaled_minmax[col] for col in numerical_cols[:5]], labels=numerical_cols[:5])
axes[2].set_title('MinMaxScaler (Range 0-1)', fontweight='bold')
axes[2].set_ylabel('Scaled Value')
axes[2].tick_params(axis='x', rotation=45)
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("SCALING SUMMARY")
print("="*80)
print(f"StandardScaler - Mean: {df_scaled_standard.mean().mean():.4f}, Std: {df_scaled_standard.std().mean():.4f}")
print(f"MinMaxScaler - Min: {df_scaled_minmax.min().min():.4f}, Max: {df_scaled_minmax.max().max():.4f}")
```
